```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#Download data from Github
df1=pd.read_csv('https://github.com/RicardoMello92/Python-Restaurant-Order/blob/main/Restaurant%20Dataset%20Kaggle/restaurant1-orders.csv?raw=True')
df2=pd.read_csv('https://github.com/RicardoMello92/Python-Restaurant-Order/blob/main/Restaurant%20Dataset%20Kaggle/restaurant2-orders.csv?raw=True')
```

```python
df1.info()
df1.head()
```

```python
df2.info()
df2.head()
```

```python
#Convert String to Datetime

df1['Order Date']=pd.to_datetime(df1['Order Date'],format='mixed')
df1['Order Number']=df1['Order Number'].astype('str')
df1.info()
df1.head()

df2['Order Date']=pd.to_datetime(df2['Order Date'],format='mixed')
df2['Order ID']=df2['Order ID'].astype('str')
df2.info()
df2.head()
```

```python
#Extract info from Order Date

df1['Date']=df1['Order Date'].dt.date
df1['Day of Week']=df1['Order Date'].dt.day_name()
df1['Month']=df1['Order Date'].dt.month
df1['Year']=df1['Order Date'].dt.year
df1['Hour']=df1['Order Date'].dt.hour
df1.head()
```

```python
#Extract info from Order Date

df2['Date']=df2['Order Date'].dt.date
df2['Day of Week']=df2['Order Date'].dt.day_name()
df2['Month']=df2['Order Date'].dt.month
df2['Year']=df2['Order Date'].dt.year
df2['Hour']=df2['Order Date'].dt.hour
df2.head()
```

```python
#Add Total Price
df1['Total Price']=df1['Quantity']*df1['Product Price']
df1
```

```python
#Add Total Price

df2['Total Price']=df2['Quantity']*df2['Product Price']
df2
```

```python
#Drop duplicates values

df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
```

```python
df1.info()

df2.info()
```

```python
#Generate descriptive statistics.

df1.describe(include='all')
```

```python
#Generate descriptive statistics.

df2.describe(include='all')
```

#Orders by Date

```python
df_date=df1[['Order Number','Date']].groupby(['Order Number','Date']).count().reset_index()
df_date.head()
```

```python
df_date.duplicated().sum()
```

```python
df_date.drop_duplicates(inplace=True)
```

```python
df_date=df_date[['Order Number','Date']].groupby('Date').count().sort_values(by='Order Number', ascending=False)
df_date
```

#Orders by Month

```python
df_month=df1[['Order Number','Month']].value_counts().to_frame().reset_index()
df_month.head()
```

```python
df_month=df_month[['Order Number','Month']].groupby('Month').count()
df_month
```

```python
#Checking results using January - 1072 Orders
df1[df1['Month']==1][['Order Number']].drop_duplicates()
```

```python
df_month.plot(kind='bar', title='Orders by Month')
```

Higher number of orders in July

Lowest number of orders in November

Highest demand period is from March to July (December is also a relevant month out of this interval)

#Orders by Year

```python
df_year=df1[['Order Number','Year']].value_counts().reset_index()[['Year','count']].groupby('Year').count()
df_year
```

```python
df_year.plot(kind='bar', title='Orders by Year')
```

Yearly increase in orders quantity from 2015 to 2018 with a drop in 2019

Highest value in 2018

#Orders by Hour

```python
df_hour=df1[['Order Number','Hour']].value_counts().reset_index()[['Hour','count']].groupby('Hour').count()
df_hour
```

```python
df_hour.plot(kind='bar', title='Orders by Hour')
```

Highest demand from 17h to 20h

#Orders by day of Week

```python
df_dow=pd.DataFrame(df1[['Order Number','Day of Week']].value_counts()).reset_index()[['Day of Week','count']].groupby('Day of Week').count()
df_dow = df_dow.reindex(['Monday','Tuesday','Wednesday', 'Thursday','Friday','Saturday','Sunday'], axis=0)
df_dow
```

```python
df_dow.plot(kind='bar', title='Orders by day of week')
```

Highest demand on Saturday

Top 3 days of week are Saturday, Friday and Sunday

#Subplots Visualization

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2, figsize=(20,10), tight_layout=True)

df_dow.plot(kind='bar', title='Orders by day of week',ax=ax[0,0])

df_month.plot(kind='bar', title='Orders by Month', ax=ax[0,1])

df_year.plot(kind='bar', title='Orders by Year', ax=ax[1,0])

df_hour.plot(kind='bar', title='Orders by Hour', ax=ax[1,1])
```

#Same analysis for Total Revenue

```python
df1_dow_rev=df1[['Day of Week','Total Price']].groupby('Day of Week').sum()
df1_dow_rev = df1_dow_rev.reindex(['Monday','Tuesday','Wednesday', 'Thursday','Friday','Saturday','Sunday'], axis=0)

df1_year_rev=df1[['Year','Total Price']].groupby('Year').sum()

df1_month_rev=df1[['Month','Total Price']].groupby('Month').sum()

df1_hour_rev=df1[['Hour','Total Price']].groupby('Hour').sum()

```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,2, figsize=(20,10), tight_layout=True)

df1_dow_rev.plot(kind='bar', title='Total Revenue by day of week',ax=ax[0,0])


df1_month_rev.plot(kind='bar', title='Total Revenue by Month', ax=ax[0,1])

df1_year_rev.plot(kind='bar', title='Total Revenue by Year', ax=ax[1,0])

df1_hour_rev.plot(kind='bar', title='Total Revenue by Hour', ax=ax[1,1])
```

#Top 5 Item Name (Based on Revenue)

```python
df1_item=df1.pivot_table(index='Item Name', values='Total Price', columns='Year', aggfunc='sum', margins=True, margins_name='Total Revenue').sort_values(by=['Total Revenue'], ascending=False)
df1_item[1:6]
```
